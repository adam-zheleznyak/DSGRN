## Creating the network and parameter graph

```python
net_spec = 'x : ~y[2] \n y : ~x[2]'

network = MultipleThresholdNetwork(net_spec)

print("Multiple Threshold Network:")
DrawGraph(network)

print("Helper Network:")
DrawGraph(network.helper_network())

parameter_graph = MultipleThresholdParameterGraph(network)
print("Size of Multiple Threshold Parameter Graph:", parameter_graph.size())
```

## Morse Graphs

```python
Table(["Parameter Index", "Morse Graph"], 
      [ [ i, DrawGraph(MorseGraph(DomainGraph(parameter_graph.parameter(i))))] for i in range(0,parameter_graph.size())])
```

## Non-monotone Example

```python
net_spec = 'x : y[+-] \n y : ~x[+-]'

network = MultipleThresholdNetwork(net_spec)

print("Multiple Threshold Network:")
DrawGraph(network)

print("Helper Network:")
DrawGraph(network.helper_network())

parameter_graph = MultipleThresholdParameterGraph(network)
print("Size of Multiple Threshold Parameter Graph:", parameter_graph.size())
```

## Single Parameter

```python
parameter = parameter_graph.parameter(125)

domain_graph = DomainGraph(parameter)

DrawGraph(domain_graph)

morse_graph = MorseGraph(domain_graph)

DrawGraph(morse_graph)
```
